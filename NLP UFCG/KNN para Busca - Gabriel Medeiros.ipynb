```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopw = stopwords.words('portuguese')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/caiolibanio/atividade_NLP/master/csv_satiras_politicas.csv")
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

def buildCountVectorizer(docs):
  vect = CountVectorizer(stop_words=stopw)
  matrix = vect.fit_transform(docs)
  return matrix, vect


def buildTFIDFVectorizer(docs):
  vect = TfidfVectorizer(stop_words=stopw)
  matrix = vect.fit_transform(docs)
  return matrix, vect

def top_words(title):
  idx = df.index[df['title'] == title]
  row = bow.getrow(idx[0])
  weights = []
  key_list = list(vect.vocabulary_.keys())
  val_list = list(vect.vocabulary_.values())
  for i in row.nonzero()[1]:
    position = val_list.index(i)
    weights.append((key_list[position], row.toarray()[0][i]))
  weights = sorted(weights, key = lambda x:x[1], reverse=True)
  return weights
```

```python
from sklearn.neighbors import NearestNeighbors

def getNearestNeighborsTitle(k, matrix, title):
  idx = df.index[df['title'] == title]
  query = matrix.getrow(idx[0])
  neigh = NearestNeighbors(n_neighbors=k)
  neigh.fit(matrix)
  knn = neigh.kneighbors(query,k)
  return knn

def getNearestNeighborsText(k, matrix, text):
  idx = df.index[df['text'] == text]
  query = matrix.getrow(idx[0])
  neigh = NearestNeighbors(n_neighbors=k)
  neigh.fit(matrix)
  knn = neigh.kneighbors(query,k)
  return knn
```
